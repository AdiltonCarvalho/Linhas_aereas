##1\. Airline Passenger Satisfaction



Customer satisfaction scores from 120,000+ airline passengers, including additional information about each passenger, their flight, and type of travel, as well as ther evaluation of different factors like cleanliness, comfort, service, and overall experience.

##2\. Importação de bibliotecas

```python
#importando bibliotecas
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

##3\. Carga de dados

```python
#Leitura da base de dados em um dataframe pandas
df = pd.read_csv('./airline_passenger_satisfaction.csv')
```

##4\. Exploração dos dados

#4.1\. Análise da solidez dos dados

```python
#Análise dos tipos de dados por coluna
df.dtypes
```

```python
#Análise de atributos categóricos
df.select_dtypes('object').describe().transpose()
```

```python
# Análise de atributos numéricos
df.select_dtypes('number').describe().transpose()
```

#5\. Visualização dos dados

```python
#Análise de passageiros por tipo de viagem e classe
plt.figure(figsize = (8, 3))
sns.set_theme(style = 'darkgrid')
grafic_type_travel = sns.displot(data = df, x = 'Type of Travel', hue = 'Gender', col = 'Class', bins = 2)
plt.show(grafic_type_travel)
```

```python
#Análise de passageiros por idade e classe
plt.figure(figsize = (15, 3))
sns.set_theme(style = 'darkgrid')
grafic_class_age = sns.displot(data = df, x = 'Age', hue = 'Gender', col = 'Class')
plt.show(grafic_class_age)
```

```python
#Análise de passageiros por classe de vôo e gênero
plt.figure(figsize = (5, 4))
sns.set_theme(style = 'darkgrid')
grafic_class_travel = sns.countplot(data = df, x = 'Class', hue = 'Gender')
grafic_class_travel.set(title = 'Passengers by class of travel', xlabel = '', ylabel = '')
plt.show(grafic_class_travel)
```

```python
#Análise de passageiros por recorrência
plt.figure(figsize = (5, 4))
sns.set_theme(style = 'darkgrid')
grafic_recurrence = sns.countplot(data = df, x = 'Customer Type', hue = 'Gender')
grafic_recurrence.set(title = 'Passenger recurrence', ylabel = '')
plt.show(grafic_recurrence)
```

```python
#Análise de distância de vôo por classe
plt.figure(figsize = (10, 3))
sns.set_theme(style = 'darkgrid')
grafic_fligth_distance = sns.histplot(data = df, x = 'Flight Distance', hue = 'Class')
grafic_fligth_distance.set(title = 'Flight distance by class')
plt.show(grafic_fligth_distance)
```

```python
#Análise de passageiros por idade e classe
plt.figure(figsize = (15, 3))
sns.set_theme(style = 'darkgrid')
grafic_class_age = sns.displot(data = df, x = 'Age', hue = 'Gender', col = 'Class')
grafic_class_age.set(ylabel = '')
plt.show(grafic_class_age)
```
